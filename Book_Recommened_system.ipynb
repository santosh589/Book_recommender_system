```python
## Popularity based  recommendation system
```

```python
# will consider top 50 books whose ratings got minimum 250 votes
# merge ratings and books

ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.head()
```

```python
number_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
number_rating_df.rename(columns={'Book-Rating': 'number_ratings'}, inplace=True)
number_rating_df.dtypes
```

```python
# Calculate average rating for each book title
average_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the 'Book-Rating' column to 'average_ratings'
average_rating_df.rename(columns={'Book-Rating': 'average_ratings'}, inplace=True)

# Print the resulting DataFrame
print(average_rating_df)
```

```python

## merging number_rating_df and average_rating_df

Popular_df= number_rating_df.merge(average_rating_df, on='Book-Title')
Popular_df.columns
```

```python
## Will take only books which have minimum 250 number of ratings
Popular_df.columns
Popular_df =  Popular_df[Popular_df['number_ratings']>=250].sort_values('average_ratings', ascending=False)
Popular_df.shape
```

```python
## Final Popular_df is ready
Popular_df = Popular_df.merge(books, on= 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author', 'Image-URL-M', 'number_ratings', 'average_ratings']]
```

```python
Popular_df.shape
```

```python
### NOW WE LOOK FOR COLLOBORATIVE FILTERING RECOMMENDER SYSTEM

## will create df with books as rows, and user as columns
## will take user rating who rated more than 200 books (user>200)
## will take popular 50 books only
```

```python
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
read_books_users= x[x].index ## who voted 200 books more
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(read_books_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index= 'Book-Title', columns= 'User-ID',values= 'Book-Rating' )
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
## to get ner by ratings books when user search for a books to get recommended near by ratings 4 books
## need to calculate euclidian distance

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[705]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('The Notebook')
```
